### Introduction About the Data :

**The dataset** The goal is to predict `price` of given diamond (Regression Analysis).

There are 10 independent variables (including `id`):

* `id` : unique identifier of each diamond
* `carat` : Carat (ct.) refers to the unique unit of weight measurement used exclusively to weigh gemstones and diamonds.
* `cut` : Quality of Diamond Cut
* `color` : Color of Diamond
* `clarity` : Diamond clarity is a measure of the purity and rarity of the stone, graded by the visibility of these characteristics under 10-power magnification.
* `depth` : The depth of diamond is its height (in millimeters) measured from the culet (bottom tip) to the table (flat, top surface)
* `table` : A diamond's table is the facet which can be seen when the stone is viewed face up.
* `x` : Diamond X dimension
* `y` : Diamond Y dimension
* `x` : Diamond Z dimension

Target variable:
* `price`: Price of the given Diamond.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Create output directory to save images
output_dir = "images"
os.makedirs(output_dir, exist_ok=True)

# 1. **Price Distribution**: Histogram showing skewed price distribution
plt.figure(figsize=(8, 6))
sns.histplot(data["price"], kde=True, color="blue")
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.savefig(f"{output_dir}/price_distribution.png", dpi=300, bbox_inches="tight")
plt.close()

# 2. **Carat vs. Price**: Scatter plot showing strong positive correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="carat", y="price", alpha=0.5)
plt.title("Carat vs. Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.savefig(f"{output_dir}/carat_vs_price.png", dpi=300, bbox_inches="tight")
plt.close()

# 3. **Cut Quality vs. Price**: Boxplot to highlight price differences by cut categories
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x="cut", y="price", palette="Set2")
plt.title("Cut Quality vs. Price")
plt.xlabel("Cut Quality")
plt.ylabel("Price")
plt.savefig(f"{output_dir}/cut_quality_vs_price.png", dpi=300, bbox_inches="tight")
plt.close()

# 4. **Correlation Heatmap**: Relationship between features and price
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig(f"{output_dir}/correlation_heatmap.png", dpi=300, bbox_inches="tight")
plt.close()

print("All plots have been saved successfully!")
```

```python
# it is a oridinal encoding
cut_map={"Fair":1,"Good":2,"Very Good":3,"Premium":4,"Ideal":5}
clarity_map = {"I1":1,"SI2":2 ,"SI1":3 ,"VS2":4 , "VS1":5 , "VVS2":6 , "VVS1":7 ,"IF":8}
color_map = {"D":1 ,"E":2 ,"F":3 , "G":4 ,"H":5 , "I":6, "J":7}
```

```python
data["cut"]=data["cut"].map(cut_map)
data["clarity"]=data["clarity"].map(clarity_map)
data["color"]=data["color"].map(color_map)
```
